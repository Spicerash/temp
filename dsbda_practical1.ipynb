```python
df['gender'].replace(['M','F'],[0,1],inplace=True)
```

```python

# Ordinal encoder using Scikit-learn
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
df[['raisedhands']]=enc.fit_transform(df[['raisedhands']])
```

```python
print('After converting categorical variable to numeric variable:\n')
print(df.head().T)
```

```python

df['raisedhands']=(df['raisedhands']-df['raisedhands'].min())/(df['raisedhands'].max()-df['raisedhands'].min())
# (x - min value into that column)/(max value - min value)
```

```python
from sklearn.preprocessing import MaxAbsScaler
abs_scaler=MaxAbsScaler()
df[['VisITedResources ']]=abs_scaler.fit_transform(df[['VisITedResources']])
```
