```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
df = pd.read_csv('Social_Network_Ads.csv')
df.head()
```

```python
df.describe()
```

```python
plt.scatter(df['Age'], df['Purchased'])
plt.xlabel("age")
plt.ylabel("purchased")
plt.show()
# Divide the data to training set and test set
X_train, X_test, y_train, y_test = train_test_split(df['Age'], df['Purchased'],
test_size=0.20)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
# ravel function returns the continuous flattened 1D Array
model.fit(X_train.values.reshape(-1, 1), y_train.values.reshape(-1, 1).ravel())
```

```python
# make prediction
y_pred_sk = model.predict(X_test.values.reshape(-1, 1))
plt.clf()
plt.scatter(X_test, y_test)
plt.scatter(X_test, y_pred_sk, c="red")
```

```python
plt.xlabel("age")
plt.ylabel("purchased")
plt.show()
# Accuracy
print(f"Accuracy = {model.score(X_test.values.reshape(-1, 1),y_test.values.reshape(-1, 1))}")
```

```python
from sklearn.metrics import confusion_matrix
```

```python
tn,fp,fn,tp=confusion_matrix(y_test,y_pred_sk).ravel()
print("True Negatives:",tn)
print("False Positives:",fp)
print("False Negatives:",fn)
print("True Positives:",tp)
```

```python
#Accuracy
Accuracy=(tn+tp)*100/(tp+tn+fp+fn)
print("Accuracy {:0.2f}%:".format(Accuracy))
```

```python
#Precision
Precision=tp/(tp+fp)
print("Precision {:0.2f}".format(Precision))
```

```python
#Recall
Recall=tp/(tp+fn)
print("Recall {:0.2f}".format(Recall))
```

```python
#Error rate
err=(fp+fn)/(tp+tn+fn+fp)
print("Error rate {:0.2f}".format(err))
```
